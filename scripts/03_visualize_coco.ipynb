```python
# Visualize labels from COCO
import os
import json
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Set file path for COCO JSON file
coco_path = '/home/cara/ct_classifier/coco_ground_crop.json'

# Set path to image folder
image_folder = '/home/cara/ct_classifier/data/ground_cropped_images'

# Load the new COCO JSON file
with open(coco_path, 'r') as f:
    new_coco_data = json.load(f)
```

```python
# Choose an image ID to visualize:
image_to_plot = "21808-3__21808-3-G__2021-05-10__08-24-08(1)_fm"
#image_to_plot = '20779-2__20779-2-G__2021-04-17__17-16-39(1)_dp'

# Find the image info from the COCO JSON
image_info = next(image for image in new_coco_data['images'] if image['id'] == image_to_plot)
annotations = [annotation for annotation in new_coco_data['annotations'] if annotation['image_id'] == image_to_plot]

# Load and display the image
image_path = os.path.join(image_folder, image_info['file_name'])
image = plt.imread(image_path)

# Find the class name 
category_id_to_find = annotations[0]['category_id'] 
category_name = None
for category in new_coco_data['categories']:
    if category['id'] == category_id_to_find:
        category_name = category['name']
        break  # Once found, exit the loop

# Choose position for the text (upper left ish)
x_pos = 30  
y_pos = 30  
    
# Plot
plt.imshow(image)
plt.text(x_pos, y_pos, category_name, color='r', fontsize=12, backgroundcolor='white', verticalalignment='top')
```
